```python
import networkx as nx
import matplotlib.pyplot as plt
import pandas as pd

#Create a Directed Graph for the main tree
DG=nx.DiGraph()

#final recipes list
recipesL = []

#add recipes
#Mandrake Soup
listm = ["Mandrake", "Stick"]
DG.add_nodes_from (listm)
for i in listm:
  DG.add_edge(i, "Mandrake Soup")

#Bacon and Eggs
listbe = ["Morsel", "Monster Meat", "Egg", "Tallbird Egg", "Stick"]
DG.add_nodes_from (listbe)
for i in listbe:
  DG.add_edge(i, "Bacon and Eggs")

#DragonPie
listdp = ["Dragon Fruit", "Stick"]
DG.add_nodes_from (listdp)
for i in listdp:
  DG.add_edge(i, "DragonPie")

#Meaty Stew
listms = ["Monster Meat", "Meat", "Morsel", "Red Cap", "Egg"]
DG.add_nodes_from (listms)
for i in listms:
  DG.add_edge(i, "Meaty Stew")

#Honey Ham
listhh = ["Monster Meat", "Meat", "Morsel", "Honey"]
DG.add_nodes_from (listhh)
for i in listhh:
  DG.add_edge(i, "Honey Ham")

  #Turkey Dinner
listtd = ["Monster Meat", "Drumstick", "Cranberries"]
DG.add_nodes_from (listtd)
for i in listtd:
  DG.add_edge(i, "Turkey Dinner")

  #Meatballs
listmb = ["Monster Meat", "Morsel",
          "Red Cap", "Egg", "Cranberries"]
DG.add_nodes_from (listmb)
for i in listmb:
  DG.add_edge(i, "Meatballs")

  #Froggle Bunwich
listfb = ["Frog Legs", "Carrot", "Stick", "Red Cap"]
DG.add_nodes_from (listfb)
for i in listfb:
  DG.add_edge(i, "Froggle Bunwich")

  #Butter Muffin
listbm = ["Butterfly Wings", "Carrot", "Stick", "Red Cap"]
DG.add_nodes_from (listbm)
for i in listbm:
  DG.add_edge(i, "Butter Muffin")

#Waffles
listw = ["Butter", "Stick", "Egg", "Cranberries"]
DG.add_nodes_from (listw)
for i in listw:
  DG.add_edge(i, "Waffles")

#Fishsticks
listw = ["Monster Meat", "Stick", "Meat", "Fish", "Fish Morsel", "Honey"]
DG.add_nodes_from (listw)
for i in listw:
  DG.add_edge(i, "Fishsticks")

#Pierogi
listp = ["Monster Meat", "Red Cap", "Morsel", "Egg", "Lichen"]
DG.add_nodes_from (listp)
for i in listp:
  DG.add_edge(i, "Pierogi")

#Trail Mix
listtm = ["Birchnut", "Berries", "Stick", "Durian"]
DG.add_nodes_from (listtm)
for i in listtm:
  DG.add_edge(i, "Trail Mix")

#Honey Nuggets
listhn = ["Monster Meat", "Honey"]
DG.add_nodes_from (listhn)
for i in listhn:
  DG.add_edge(i, "Trail Mix")

#Fruit Medley
listfm = ["Durian", "Banana",  "Watermelon", "Stick"]
DG.add_nodes_from (listfm)
for i in listfm:
  DG.add_edge(i, "Fruit Medley")

#Fruit Medley
listfm = ["Durian", "Banana",  "Stick"]
DG.add_nodes_from (listfm)
for i in listfm:
  DG.add_edge(i, "Fruit Medley")

#Taffy
listt = ["Honey"]
DG.add_nodes_from (listt)
for i in listt:
  DG.add_edge(i, "Taffy")

#Jelly Salad
listjs = ["Leafy Meat", "Honey", "Honeycomb"]
DG.add_nodes_from (listjs)
for i in listjs:
  DG.add_edge(i, "Jelly Salad")

#Leafy Meatloaf
listlm = ["Leafy Meat", "Stick"]
DG.add_nodes_from (listlm)
for i in listlm:
  DG.add_edge(i, "Leafy Meatloaf")

# - - - #

print("Dont Starve Crock Pot Ingredients:")
user_input = input() #takes user inputed ingredient
my_list = user_input.split(", ") #split inputs into a list of items

#finds the successors of the entered ingredient
for i in my_list: #iterate through user inputs
    r = list(DG.successors(i)) #make all recipies/sucessors into a list (r)
    recipesL.append(r) #make a list (recipes) of lists (each recipe/sucessor of input)



recipes = [pd.Series(r) for r in recipesL]

#inner merges depening on number of items inputed and saves final merge to var final
if len(recipes) == 4:
  first = pd.merge(recipes[0].to_frame(), recipes[1].to_frame(), how='inner', on=0)
  second = pd.merge(recipes[2].to_frame(), recipes[3].to_frame(), how='inner', on=0)
  final = pd.merge(first, second, how='inner', on=0)

elif len(recipes) == 3:
  first = pd.merge(recipes[0].to_frame(), recipes[1].to_frame(), how='inner', on=0)
  final = pd.merge(first, recipes[2], how='inner', on=0)

elif len(recipes) == 2:
  final = pd.merge(recipes[0].to_frame(), recipes[1].to_frame(), how='inner', on=0)

elif len(recipes) == 1:
  final = recipes[0].to_frame()

#output
print()
print("Available recipes using", user_input, ":",)
print(final)


#draw graph - uncomment two lines below to visualize graph
#nx.draw_circular(DG, with_labels =True, node_size=2000, font_size=10, arrowsize = 20, node_color="#FFFFFF" , edge_color="#AEAEAE",bbox=dict(facecolor="#add8e6", edgecolor="white", boxstyle="round, pad=0.2"))
#plt.savefig("basicgraph2.png")
```
